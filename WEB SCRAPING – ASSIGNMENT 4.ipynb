```python
# importing necessary libreries.
import selenium 
import pandas as pd
import warnings
from selenium import webdriver
from selenium.webdriver.common.by import By
warnings.filterwarnings('ignore')
import time
```

```python
# opening = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
driver = webdriver.Chrome()
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos ")
time.sleep(2)
driver.maximize_window()
```

```python
#creating emepty list to store details.
rank_list = []
name_list=[]
artist_list = []
upload_date= []
views = []

    
```

```python
# scraping the all details required
ranks_tags=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
for i in ranks_tags:
    rank=i.text
    rank_list.append(rank)
name_tags=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
for i in name_tags:
    names=i.text
    name_list.append(names)
artist_tags=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]')  
for i in artist_tags:
    artist=i.text
    artist_list.append(artist)
up_date_tags=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[5]')  
for i in up_date_tags:
    date=i.text
    upload_date.append(date)
views_tags=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]') 
for i in views_tags:
    view=i.text
    views.append(view)
    
    
    
    
    
 
    
```

```python
print(len(rank_list),len(name_list),len(artist_list),len(upload_date),len(views))
```

```python
Book_name=[]
Author_name=[] 
Volumes_sold=[] 
Publisher=[] 
Genre=[] 
books_tags=driver.find_elements(By.XPATH,'//table[@class="in-article sortable"]/tbody/tr/td[2]')
for i in books_tags:
    bookes=i.text
    Book_name.append(bookes)
author_tags=driver.find_elements(By.XPATH,'//table[@class="in-article sortable"]/tbody/tr/td[3]')
for i in author_tags:
    authors=i.text
    Author_name.append(authors)
volume_tags=driver.find_elements(By.XPATH,'//table[@class="in-article sortable"]/tbody/tr/td[4]')
for i in volume_tags:
    volume=i.text
    Volumes_sold.append(volume)
publisher_tags=driver.find_elements(By.XPATH,'//table[@class="in-article sortable"]/tbody/tr/td[5]')
for i in publisher_tags:
    publisher=i.text
    Publisher.append(publisher)
genre_tags=driver.find_elements(By.XPATH,'//table[@class="in-article sortable"]/tbody/tr/td[6]')
for i in genre_tags:
    genre=i.text
    Genre.append(genre)
```

```python
print(len(Book_name),len(Author_name),len(Volumes_sold),len(Publisher),len(Genre))
```
